# Visualization

```python
state_counts = my_data['state'].value_counts()
plt.figure(figsize=(10,6))
state_counts.plot(kind ='bar', color ='skyblue')
plt.title('Frequency')
plt.xlabel('States')
plt.ylabel('Count')
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
avg_pollutants = my_data.groupby('pollutant_id')['pollutant_avg'].mean()
plt.figure(figsize=(10,6))
avg_pollutants.plot(kind='bar', color='skyblue')
plt.title('Average Pollutants Across the Day')
plt.xlabel('Pollutant Type')
plt.ylabel('Average')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
avg_pollutants_state = my_data.groupby('state')['pollutant_avg'].mean()
plt.figure(figsize=(12,6))
avg_pollutants_state.plot(kind='bar', color='green')
plt.title('Average Pollutants Across the States')
plt.xlabel('States')
plt.ylabel('Average')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
avg_pollutants_state_type = my_data.groupby(['state','pollutant_id'])['pollutant_avg'].mean()
plt.figure(figsize=(12,6))
avg_pollutants_state_type.plot(kind='line', color='green')
plt.title('Different Pollutant Levels Across the States')
plt.xlabel('States')
plt.ylabel('Average')

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
max_pollutants_state = my_data.groupby('state')['pollutant_max'].mean()
plt.figure(figsize=(12,6))
max_pollutants_state.plot(kind='bar', color='orange')
plt.title('Maximum Pollutants Across the States')
plt.xlabel('States')
plt.ylabel('Maximum')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
min_pollutants_state = my_data.groupby('state')['pollutant_min'].mean()
plt.figure(figsize=(12,6))
min_pollutants_state.plot(kind='line', color='purple')
plt.title('Minimum Pollutants Across the States')
plt.xlabel('States')
plt.ylabel('Minimum')
plt.xticks(rotation=45, ha='right')
plt.show()
```
